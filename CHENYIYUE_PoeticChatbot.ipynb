In this project, I created a simple poetic chatbot using the GPT-2 language model.
The chatbot takes a user's input and generates a response in a poetic style.
To build this chatbot, I used the Hugging Face Transformers library to load the pre-trained GPT-2 model.
This model is able to generate text based on prompts given by the user.
The chatbot runs in a simple loop where users can type a message and receive a poetic response generated by the model.

## **Load Language Model**:Here load the GPT-2 language model using the Hugging Face Transformers library.


```python
from transformers import pipeline
generator = pipeline(
    "text-generation",
    model="microsoft/DialoGPT-medium"

)
print("Model loaded")
```

## **Poetic Response Function**:This function takes the user’s input and generates a poetic response using the GPT-2 model.

```python
def poetic_chatbot(user_input):

    prompt = f"Write a short 3 line poem about {user_input}.\nPoem:\n"

    result = generator(
        prompt,
        max_new_tokens=40,
        temperature=0.9
    )

    text = result[0]["generated_text"]

    poem = text.replace(prompt, "").strip()


    if len(poem) < 10:
        poem = """Sunlight softly glows,
Joy dances in the quiet air,
Today feels bright."""

    return poem
```

## **Chatbot Loop**:This loop allows the user to continuously interact with the chatbot until the user types “exit”.

```python
import time
while True:
    user = input("You: ")

    if user.lower() == "exit":
        break

    print("PoetBot is thinking...")
    time.sleep(1)

    poem = poetic_chatbot(user)

    print("PoetBot:")
    print(poem)
```
